```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator


log_list = []
for line in open("log_best.txt"): 
    lr, wd, order, a, b, c, plus, f1 = line.split(' ')
#     print(line)
    log_list.append([lr, wd, order, plus, float(f1)])

log_list = np.array(log_list)
```

```python
log_list_no = log_list[log_list[:,3]=='0']
log_list_plus = log_list[log_list[:,3]=='1']
```

```python
print(max(log_list_plus[:,-1].astype(float)))
print(max(log_list_no[:,-1].astype(float)))
```

```python
sns.violinplot(log_list[:,3].astype(int), log_list[:,4].astype(float), alpha=.2)
# plt.ylim(0.5,0.9)
plt.xlabel('0 denote SGC, 1 denote SGC++', fontsize=15)
plt.ylabel('f1 score ',fontsize=15)

plt.savefig('./overall.pdf' ,bbox_inches='tight',pad_inches=0.0)
plt.show()
```

```python
plt.subplot(1,2,1)
sns.violinplot(log_list_no[:,2].astype(int), log_list_no[:,4].astype(float), alpha=.2)
plt.ylim(0.8,0.97)
plt.xlabel('order (SGC)', fontsize=15)
plt.ylabel('f1 score ',fontsize=15)
plt.subplot(1,2,2)
sns.violinplot(log_list_plus[:,2].astype(int), log_list_plus[:,4].astype(float), alpha=.2)
plt.ylim(0.8,0.97)
plt.xlabel('order (SGC++)', fontsize=15)
# plt.ylabel('f1 score ',fontsize=15)
plt.savefig('./two.pdf' ,bbox_inches='tight',pad_inches=0.0)
plt.show()
```
